```python
import pandas
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
```

```python
dataframe = pandas.read_csv("sonar.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:60].astype(float)
Y = dataset[:,60]
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

```python
model = Sequential()
model.add(Dense(60, input_dim=60, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

basic_model = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
basic_model.graph_name = 'basic_model'
```

```python
model = Sequential()
model.add(Dense(30, input_dim=30, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

X30 = dataset[:,0:30].astype(float)

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

half_model = model.fit(X30, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
half_model.graph_name = 'half_model'
```

```python
model = Sequential()
model.add(Dense(30, input_dim=30, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

X30_last = dataset[:,30:60].astype(float)

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

half_model_last = model.fit(X30_last, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
half_model_last.graph_name = 'half_model_last'
```

```python
model = Sequential()
model.add(Dense(60, input_dim=60, kernel_initializer='normal', activation='relu'))
model.add(Dense(15, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

hidden_model = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
hidden_model.graph_name = 'hidden_model'
```

```python

import matplotlib.pyplot as plt

models = [basic_model, half_model, hidden_model, half_model_last]

for model in models:
  # График ошибки
  plt.plot(model.history['loss'], label='train_loss ' + model.graph_name)
  plt.plot(model.history['val_loss'], label='val_loss ' + model.graph_name)
  plt.xlabel('Эпоха')
  plt.ylabel('Потери')
  plt.legend()
  plt.title('График функции потерь')
plt.show()

for model in models:
  # График точности
  plt.plot(model.history['accuracy'], label='train_accuracy ' + model.graph_name)
  plt.plot(model.history['val_accuracy'], label='val_accuracy ' + model.graph_name)
  plt.xlabel('Эпоха')
  plt.ylabel('Точность')
  plt.legend()
  plt.title('График точности')
plt.show()
```

Уменьшение числа входных нейронов с 60 до 30 почти не изменило результаты, вероятнее всего потому что для небольшого числа сэмплов разница между 30ю и 60ю измерениями небольшая. Первые 30 измерений более важные, чем последние 30

Добавление промежуточного слоя позволило модели давать более точные предсказания, хотя тренировка шла дольше
